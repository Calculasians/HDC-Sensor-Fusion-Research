```python
def rule90(a, lent):
    right = format(int(bin(a >> 1), 2), '#0{l}b'.format(l=lent+2))
    left = format(int(bin(a << 1), 2), '#0{l}b'.format(l=lent+2))
    left = "0b{l}".format(l=left[len(left)-lent:len(left)])
    return format(int(bin(int(left, 2) ^ int(right, 2)),2), '#0{l}b'.format(l=lent+2))
```

```python
def foldedrule90(a, lent, ffactor):
    fresults = []
    flist = [a[ffactor*i:ffactor*(i+1)] for i in range(len(a)//ffactor)]

    for f in flist:
        f = "0b{f}".format(f=f)
        r90 = rule90(int(f, 2), ffactor)
        fresults.append(r90[2:len(r90)])

    result = ""
    for f in fresults:
        result += f

    return result
```
